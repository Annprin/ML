`````python
def train(model, X_train, y_train, X_test, y_test):
    time_fit = time.time()
    model.fit(X_train, y_train)
    time_fit = time.time() - time_fit
    y_pred = model.predict(X_test)
    mae = np.mean(np.abs(y_pred - y_test))
    return time_fit, mae
`````

`````python
times = []
maes = []
random_tree_regressor = RandomForestRegressor(n_estimators=100, warm_start=True, random_state=42)
time_fit, mae = train(random_tree_regressor, X_train, y_train, X_test, y_test)
times.append(time_fit)
maes.append(mae)
while(random_tree_regressor.n_estimators < 1000):
    random_tree_regressor.n_estimators += 100
    time_fit, mae = train(random_tree_regressor, X_train, y_train, X_test, y_test)
    times.append(time_fit)
    maes.append(mae)
`````

`````python
class PotentialTransformer:
    """
    A potential transformer.

    This class is used to convert the potential's 2d matrix to 1d vector of features.
    """

    def fit(self, x, y):
        """
        Build the transformer on the training set.
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: trained transformer
        """
        return self

    def fit_transform(self, x, y):
        """
        Build the transformer on the training set and return the transformed dataset (1d vectors).
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: transformed potentials (list of 1d vectors)
        """
        return self.transform(x)

    def transform(self, x):
        """
        Transform the list of potential's 2d matrices with the trained transformer.
        :param x: list of potential's 2d matrices
        :return: transformed potentials (list of 1d vectors)
        """
        return x.reshape((x.shape[0], -1))

SVRTransform = Pipeline([
    ('scaler', PotentialTransformer()),
    ('model', LinearSVR())
])

RFRTransform = Pipeline([
    ('scaler', PotentialTransformer()),
    ('model', RandomForestRegressor())
])
`````

`````python
class SpecialTransformer:
    """
    A potential transformer.

    This class is used to convert the potential's 2d matrix to 1d vector of features.
    """

    def fit(self, x, y):
        """
        Build the transformer on the training set.
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: trained transformer
        """
        return self

    def fit_transform(self, x, y):
        """
        Build the transformer on the training set and return the transformed dataset (1d vectors).
        :param x: list of potential's 2d matrices
        :param y: target values (can be ignored)
        :return: transformed potentials (list of 1d vectors)
        """
        return self.transform(x)

    def transform(self, x):
        """
        Transform the list of potential's 2d matrices with the trained transformer.
        :param x: list of potential's 2d matrices
        :return: transformed potentials (list of 1d vectors)
        """
        for i, image in enumerate(x):
            # print(image.shape)
            min_value = np.min(image)
            eps = 0.1
            min_coords = np.where(image == min_value)
            coords = np.where(np.abs(image - min_value) > eps)
            # image[coords] = 20
            center_of_mass = (np.mean(min_coords[0]), np.mean(min_coords[1]))
            center_of_mass = (int(round(center_of_mass[0])), int(round(center_of_mass[1])))
            # print(" :", min_coords)
            #  
            image_center = (128, 128)  #   256x256
            #  
            shift = (image_center[0] - center_of_mass[0], image_center[1] - center_of_mass[1])
            #  
            image = np.roll(image, shift, axis=(0, 1))

            #     20
            if shift[0] > 0:
                image[:shift[0], :] = 20  #  
            else:
                image[shift[0]:, :] = 20  #  

            if shift[1] > 0:
                image[:, :shift[1]] = 20  #  
            else:
                image[:, shift[1]:] = 20  #  
            x[i] = image.copy()
        return x.reshape((x.shape[0], -1))
        # return shifted_image.reshape((shifted_image.shape[0], -1))

RFRTransform = Pipeline([
    ('scaler', SpecialTransformer()),
    ('model', RandomForestRegressor(n_estimators=10, random_state=42))
])
`````
